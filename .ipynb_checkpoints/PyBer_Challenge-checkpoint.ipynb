# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = pyber_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = total_fares / total_rides
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fares / total_drivers
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.concat({'Total Rides': total_rides,'Total Drivers': total_drivers,'Total Fares': total_fares,'Average Fare per Ride': average_fare_per_ride,'Average Fare per Driver': average_fare_per_driver}, axis=1)
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pd.options.display.float_format = '${:,.2f}'.format
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].round(decimals = 2)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].round(decimals = 2)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].round(decimals = 2)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df.head()
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
indices = ["date", "type"]
sum_of_fares_df = pyber_data_df.groupby(indices).sum()[["fare"]]
sum_of_fares_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_of_fares_df = sum_of_fares_df.reset_index()
sum_of_fares_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_of_fares_pivot = sum_of_fares_df.pivot(index="date", columns="type", values="fare")
sum_of_fares_pivot.head(10)
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-28'.
new_sum_of_fares_df = sum_of_fares_pivot.loc[(sum_of_fares_pivot.index >= "2019-01-01") & (sum_of_fares_pivot.index < "2019-04-29")]
new_sum_of_fares_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
new_sum_of_fares_df.index = pd.to_datetime(new_sum_of_fares_df.index)
new_sum_of_fares_df
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
new_sum_of_fares_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_of_fares_week = new_sum_of_fares_df.resample("W").sum()
sum_of_fares_week.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(20, 6))
sum_of_fares_week.plot(kind='line', ax=ax, color=['lightskyblue', 'lightcoral', 'gold'])
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare($USD)")
ax.set_xlabel("Date")

# Save the figure
plt.savefig("analysis/PyBer_fare_summary.png")

# Display the figure
plt.show()
```
